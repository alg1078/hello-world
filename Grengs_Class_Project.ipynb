### Introduction

Calcium (Ca) serves as a vital element as it is the most abundant mineral nutrient in Earth’s crust and serves an important role in chemical weathering especially within a forest (Likens et al., 1998; Hynicka et al., 2016). Quantifying Ca is often a challenge because of its mobility after being leached and liberated from soils (Federer et al., 1989; Bailey et al., 1996; Likens et al., 1998). As seen in the northeast, Ca depletion is often a consequence of acid rain and forest harvesting (Bailey et al., 1996; Blum et al., 2008; Green et al., 2013). The deposition of nitrogen and sulfur is also suggested to cause Ca leaching (Vadeboncoeur, 2010) along with other variables including cation exchange, mycorrhizal infections, tree frost injuries, and organic matter decomposition (Likens et al., 1998; Blum et al., 2008). Understanding how and when Ca is leached can help forest management practices improve for maintaining a healthy forest. 

In this study, two of Hubbard Brook Experimental Forest’s (HBEF) watersheds will be compared to observe when Ca concentrations are at their highest within its watershed’s soil concentrations and streamwater fluxes. Hubbard Brook is an interesting site to select because in October of 1999 (HBEF) had 45 tons of wollastonite (CaSiO3) added to one of its watersheds, specifically watershed 1 (W1) (Green et al., 2013; “Hubbard Brook Ecosystem Study”, n.d). Adding wollastonite to HB's W1 was in response to the acid deposition depleting the soil's Ca within the northeast region (Green et al., 2013). 

The primary approach was first to see what the baseline Ca concentration was to observe any specific seasons when Ca is highest without the effect of the wollastonite addition. It was assumed that Ca baseline concentrations could be found from 2012 to 2017 by finding the averages for each month during this five-year span. This will help determine if there was a specific season in which Ca is highest within its organic (O) horizon, mineral (B) horizon, and streamwater chemistry. From this plot, Ca will be further focused on a decade timescale during the wollastonite deposition to observe differences in stream water, soil, and pH concentrations. 

**Specifically, the following questions will be addressed:**

* Within a five year span, when will the O horizon's Ca concentration be highest seasonally at W1? Do the B horizon and streamwater follow suit? Will W6 show similarity with W1?

* If there is a seasonal pattern of when Ca concentrations are at it's peak in the O horizon in W1, how does Ca change for a decade after the deposition of wollasonite in October 1999 based on it's seasonality? Again will the streamwater flux, O horizon silica concentration, and W6 follow a similar trend?

* Does soil Ca concentrations and pH along with Ca fluxes and pH impact Ca's repsonse among the tested watersheds?

### Methods and Results



First, the necessary libraries including Datetime, Pandas, and Matplotlib were imported. These libraries are used throughout the entirety of the analysis. 

```python
#Importing external libraries 

from datetime import datetime
import datetime               
import pandas as pd
from matplotlib import pyplot as plt 
```

Next, inputted filenames were specified and variable names added for easier abbreivations to use for later coding.

```python
#%% Specify inputs/filenames 

filenameW1_soil = ['HBEF_1996-2017_Soil_Chem_W1.csv']
filenameW6_soil = ['HBEF_1984-2017_Soil_Chem_W6.csv']
filenameW1_stream = ['ws1_stream_monthly_flux_gHa.csv']
filenameW6_stream = ['ws6_stream_monthly_flux_gHa.csv']

#Location of study site
c1 = 'Hubbard Brook, NH' 
w1 = 'Watershed 1'
w6 = 'Watershed 6'
```

The data was imported based on the recommendations from the creators for a smooth transition into python. The imported data was modified by creating functions for both the soil solutions and streamwater chemistry datafiles. This helped streamline each watershed's data quicker. 

For the soil solutions data, a dictionary was created to create simple soil horizons. Horizons Oa, Oe, and Oae were grouped into a simple O horizon group. Along with, horizons Bh, Bhs, Bhs1, Bhs3, Bs, and Bs2 were grouped into a simple B horizon group. This data was filtered to bring in the following columns: watershed, horizon, Ca concentration, silica concentration, date, and pH. The index was set to the date. 

The streamwater chemistry data was not like the soil solutions data by not having an exact date. Thus, to start with the index was set to year. Then, the data was filtered to bring in the following columns: month, flow, Ca flux, year datetime, pH, and year-month.

Both datasets NA values set to equal -999.99, -888.88, -777.77, 'NaN', 'Ice', or 'Eqp'. 


```python
#To import data, using what HB creators recommended for smooth transition
filenameW1_soil  ="https://pasta.lternet.edu/package/data/eml/knb-lter-hbr/138/9/e531788965a5622a4ca8e664a9d06406".strip() 
filenameW1_soil  = filenameW1_soil.replace("https://","http://")
filenameW6_soil  ="https://pasta.lternet.edu/package/data/eml/knb-lter-hbr/62/16/4fa59538abe2e6ff15a35ea2de913546".strip() 
filenameW6_soil  = filenameW6_soil.replace("https://","http://")
filenameW1_stream  ="https://pasta.lternet.edu/package/data/eml/knb-lter-hbr/3/17/520d38828fe2356314e51008a5059dd4".strip() 
filenameW1_stream  = filenameW1_stream.replace("https://","http://")
filenameW6_stream  ="https://pasta.lternet.edu/package/data/eml/knb-lter-hbr/8/17/3312389e77cc5fd06bc8a7c9019de0ed".strip() 
filenameW6_stream  = filenameW6_stream.replace("https://","http://")

#Created a dictoriary before importing soils data
simple_hori = {'Oa' : 'O', 'Oe' : 'O', 'Oae' : 'O', 'Bh' : 'B', 'Bhs' : 'B', 'Bhs1' : 'B', 'Bhs3' : 'B', 'Bs' : 'B', 'Bs2' : 'B'}

#Read in soil chemistry datafile
def read_in_soils(file):
    df = pd.read_csv(file, skiprows=1, sep=",", names=["date","WS","Elev","Hor","pH","ANC","H4SiO4","Na_plus_",     
             "K_plus_","Ca2_plus_","CaFLAG", "Mg2_plus_","Alt","Alm", "Alo","NH4_plus_", "F_hyphen_",     
             "Cl_hyphen_", "SO4=","NO3_hyphen_","DIC", "DOC","TN"],
             parse_dates=['date'], na_values = [-999.99,-888.88, -777.77, 'NaN', 'Ice', 'Eqp']) 
    df=df.assign(date_datetime=pd.to_datetime(df.date,errors='coerce'))  
    df.WS=df.WS.astype('category')  
    df.Elev=df.Elev.astype('category')  
    df.Hor=df.Hor.astype('category') 
    df.pH=pd.to_numeric(df.pH,errors='coerce') 
    df.ANC=pd.to_numeric(df.ANC,errors='coerce') 
    df.H4SiO4=pd.to_numeric(df.H4SiO4,errors='coerce') 
    df.Na_plus_=pd.to_numeric(df.Na_plus_,errors='coerce') 
    df.K_plus_=pd.to_numeric(df.K_plus_,errors='coerce') 
    df.Ca2_plus_=pd.to_numeric(df.Ca2_plus_,errors='coerce')  
    df.CaFLAG=df.CaFLAG.astype('category') 
    df.Mg2_plus_=pd.to_numeric(df.Mg2_plus_,errors='coerce') 
    df.Alt=pd.to_numeric(df.Alt,errors='coerce') 
    df.Alm=pd.to_numeric(df.Alm,errors='coerce') 
    df.Alo=pd.to_numeric(df.Alo,errors='coerce') 
    df.NH4_plus_=pd.to_numeric(df.NH4_plus_,errors='coerce') 
    df.F_hyphen_=pd.to_numeric(df.F_hyphen_,errors='coerce') 
    df.Cl_hyphen_=pd.to_numeric(df.Cl_hyphen_,errors='coerce') 
    # df.SO4=pd.to_numeric(df.SO4=,errors='coerce') 
    df.NO3_hyphen_=pd.to_numeric(df.NO3_hyphen_,errors='coerce') 
    df.DIC=pd.to_numeric(df.DIC,errors='coerce') 
    df.DOC=pd.to_numeric(df.DOC,errors='coerce') 
    df.TN=pd.to_numeric(df.TN,errors='coerce') 
    df = df.set_index(['date'])
    df = df[['WS', 'Hor','Ca2_plus_', 'H4SiO4', 'date_datetime', 'pH']]
    # df = df[['WS', 'Hor','Ca2_plus_', 'H4SiO4', 'date_datetime', 'pH', 'Mg2_plus_', 'K_plus_', 'Na_plus_',
    #          'Alo', 'NH4_plus_']]
    df = df.replace({"Hor" : simple_hori})
    
    return df

#Read in streamwater chemistry datafile
def read_in_streamwater(file):
    df =pd.read_csv(file ,skiprows=1 ,sep=",",quotechar='"', names=["Year", "Month","Year_Month", "flow_mm", "Ca_flux","Mg_flux",     
            "K_flux", "Na_flux", "Al_Ferron_flux", "TMAl_flux", "OMAl_flux","Al_ICP_flux","NH4_flux","SO4_flux",     
            "NO3_flux","Cl_flux","PO4_flux","DOC_flux", "TDN_flux","DON_flux", "DIC_flux","SiO2_flux","Mn_flux",     
            "Fe_flux", "F_flux","H_flux", "pH_volwt", "SpecCond_volwt","ANC_volwt"],
            parse_dates=['Year'], na_values = [-999.99,-888.88, -777.77, 'NaN', 'Ice', 'Eqp']) 
    df=df.assign(Year_datetime=pd.to_datetime(df.Year,errors='coerce'))  
    df.Month=df.Month.astype('category')  
    df.Year_Month=df.Year_Month.astype('category') 
    df.flow_mm=pd.to_numeric(df.flow_mm,errors='coerce') 
    df.Ca_flux=pd.to_numeric(df.Ca_flux,errors='coerce') 
    df.Mg_flux=pd.to_numeric(df.Mg_flux,errors='coerce') 
    df.K_flux=pd.to_numeric(df.K_flux,errors='coerce') 
    df.Na_flux=pd.to_numeric(df.Na_flux,errors='coerce') 
    df.Al_Ferron_flux=pd.to_numeric(df.Al_Ferron_flux,errors='coerce',downcast='integer') 
    df.TMAl_flux=pd.to_numeric(df.TMAl_flux,errors='coerce') 
    df.OMAl_flux=pd.to_numeric(df.OMAl_flux,errors='coerce') 
    df.Al_ICP_flux=pd.to_numeric(df.Al_ICP_flux,errors='coerce') 
    df.NH4_flux=pd.to_numeric(df.NH4_flux,errors='coerce') 
    df.SO4_flux=pd.to_numeric(df.SO4_flux,errors='coerce') 
    df.NO3_flux=pd.to_numeric(df.NO3_flux,errors='coerce') 
    df.Cl_flux=pd.to_numeric(df.Cl_flux,errors='coerce') 
    df.PO4_flux=pd.to_numeric(df.PO4_flux,errors='coerce') 
    df.DOC_flux=pd.to_numeric(df.DOC_flux,errors='coerce') 
    df.TDN_flux=pd.to_numeric(df.TDN_flux,errors='coerce') 
    df.DON_flux=pd.to_numeric(df.DON_flux,errors='coerce') 
    df.DIC_flux=pd.to_numeric(df.DIC_flux,errors='coerce') 
    df.SiO2_flux=pd.to_numeric(df.SiO2_flux,errors='coerce') 
    df.Mn_flux=pd.to_numeric(df.Mn_flux,errors='coerce') 
    df.Fe_flux=pd.to_numeric(df.Fe_flux,errors='coerce') 
    df.F_flux=pd.to_numeric(df.F_flux,errors='coerce') 
    df.H_flux=pd.to_numeric(df.H_flux,errors='coerce') 
    df.pH_volwt=pd.to_numeric(df.pH_volwt,errors='coerce') 
    df.SpecCond_volwt=pd.to_numeric(df.SpecCond_volwt,errors='coerce') 
    df.ANC_volwt=pd.to_numeric(df.ANC_volwt,errors='coerce') 
    df = df.set_index(['Year'])
    df = df[['Month', 'flow_mm', 'Ca_flux', 'Year_datetime', 'pH_volwt', 'Year_Month']]

    return df
```

```python
#File reading for each watershed's soil and stream water chemistry

dfW1_soil = read_in_soils(filenameW1_soil)
dfW6_soil = read_in_soils(filenameW6_soil)
dfW1_stream = read_in_streamwater(filenameW1_stream)
dfW6_stream = read_in_streamwater(filenameW6_stream)
```

After all datafiles were imported, copies were made to have all dates avaiable for later analysis.

```python
#Created a copy of the original (og) dataset

dfw1_soil_og = dfW1_soil.copy()
dfw6_soil_og = dfW6_soil.copy()
dfw1_streams_og = dfW1_stream.copy()
dfw6_streams_og = dfW6_stream.copy()
```

The soil solutions datasets were first focussed on by selecting five year span that wasn't affected by CaSiO3. Selecting and trimming the dates from December 2012 to Decemeber 2017 will help establish a baseline Ca and H4SiO4 concentrations for comparison between W1 and W6. NA values were dropped since many subgroups of each horizon were combined when importing the data. 

Each watershed's year, month, and day was extracted from the index. New dataframes were created to get O and B horizons separted for timeseries plotting. From that, the mean values from each month within the five year span would calculated. Originally, separting the O and B horizons were going to be merged together, however, problems in pythgon caused the O horizon to average values. Thus, it was decided to keep each watershed's soil horizons in separate dataframes for this analyses.

```python
#Selecting a five year span to establish a baseline Ca concentration
startdate5_soil = pd.to_datetime('2012-12-14')
enddate5_soil = pd.to_datetime('2017-12-17') 
```

```python
#Triming the dataframes to data that occur between the desired baseline start date and end date
dfW1_soil = dfW1_soil[startdate5_soil:enddate5_soil]
dfW6_soil = dfW6_soil[startdate5_soil:enddate5_soil]

dfW1_soil = dfW1_soil.dropna()
dfW6_soil = dfW6_soil.dropna()

dfW1_soil['YEAR'] = dfW1_soil.index.year
dfW1_soil['DAY'] = dfW1_soil.index.dayofyear
dfW1_soil['MONTH'] = dfW1_soil.index.month

dfW6_soil['YEAR'] = dfW6_soil.index.year
dfW6_soil['DAY'] = dfW6_soil.index.dayofyear
dfW6_soil['MONTH'] = dfW6_soil.index.month

# Separting O and B horizons
dfW1_soil_O = dfW1_soil.loc[(dfW1_soil['Hor'] == 'O')]
dfW6_soil_O = dfW6_soil.loc[(dfW6_soil['Hor'] == 'O')]
dfW1_soil_B = dfW1_soil.loc[(dfW1_soil['Hor'] == 'B')]
dfW6_soil_B = dfW6_soil.loc[(dfW6_soil['Hor'] == 'B')]

#Merge them causes the O horizon to average (or do something funky) right away
# dfW1_soil = dfW1_soil_O.merge(dfW1_soil_B, how = 'inner', left_index = True, right_index = True) 
# dfW6_soil = dfW6_soil_O.merge(dfW6_soil_B, how = 'inner', left_index = True, right_index = True)

#Finding mean values for horizons
dfW1_soilO_mean = dfW1_soil_O.groupby(['MONTH']).mean()
dfW6_soilO_mean = dfW6_soil_O.groupby(['MONTH']).mean()
dfW1_soilB_mean = dfW1_soil_B.groupby(['MONTH']).mean()
dfW6_soilB_mean = dfW6_soil_B.groupby(['MONTH']).mean()
```

A similar approach was taken with the streamwater chemistry data by selecting and trimming to the years 2012 through 2017. Since a month column was already in the datafile, mean values were computed for each month within the five year span. 

```python
#Streamwater
dfW1_stream['YEAR'] = dfW1_stream.index.year
dfW6_stream['YEAR'] = dfW6_stream.index.year

#Trim years
start_yr_stm = pd.to_datetime('2012')
end_yr_stm = pd.to_datetime('2017')
dfW1_stream = dfW1_stream[start_yr_stm:end_yr_stm]
dfW6_stream = dfW6_stream[start_yr_stm:end_yr_stm]

#Finding mean values for watersheds
dfW1_stream_mean = dfW1_stream.groupby(['Month']).mean()
dfW6_stream_mean = dfW6_stream.groupby(['Month']).mean()
```

From the revised datasets, a function was created to display each watershed's baseline Ca concentrations. The first plot is showing W1 and then W6. The blue line symbolizes the organic horizon, red line is B horizon, and green is the streamwater. Note that the soil's data is units of parts per billion (ppb, umol/L) and stream water fluxes are in grams per hectare (g/ha). Months are labelled numercially along the x-axis. Please note that for the remaining plots that watershed 1 will display first and then watershed 6.

```python
def timeseriesplot(meanO, meanB, meanS):
    fig, (ax1) = plt.subplots(sharex = True)
    
    l1 = ax1.plot((meanO['Ca2_plus_']), color = 'blue', linestyle = '-', label = 'O Horizon')
    l2 = ax1.plot((meanB['Ca2_plus_']), color = 'red', linestyle = '-', label = 'B Horizon')
    
    ax2 = ax1.twinx()
    l5 = ax2.plot(meanS['Ca_flux'],'g^-',label = 'Streamwater')
    
    ax2.set_ylim(bottom = 0, top = 3000)
    ax2.set_ylabel('Ca Flux (g/ha)', color='g')
    
    fig.autofmt_xdate() 
    
    ax1.set_xlabel('Time (Month #)', color='k')
    ax1.set_ylabel('[Ca] Monthly Mean (ppb)', color='k')   

    ax1.set_ylim(bottom = 0)
    ax1.set_ylim(top = 120)

    ax1.set_title("Watershed" +' Time Series for 2012-2017')
    
    lines = l1 + l2 + l5 
    labs = [l.get_label() for l in lines]
    ax1.legend(lines, labs, loc=(1.1, 1.0))
    
    return meanO, meanB, meanS

timeseriesplot(dfW1_soilO_mean, dfW1_soilB_mean, dfW1_stream_mean)
timeseriesplot(dfW6_soilO_mean, dfW6_soilB_mean, dfW6_stream_mean)
```

It was interesting to note that both watersheds differed quite a bit in Ca concentrations and fluxes. From these plots, it was decided to focus on season where W1's highest Ca concentration in the O horizon occured during the fall months. 

Thus, in the next plot we will obsevere what was Ca's response on annual basis based during the fall months during the time when the wollasnoite was deposited in October of 1999.

To do this, a new dataframe was created from the from the soil's original dataframe and have it trimmed from dates ranging from  September 1997 to Novemeber 2007. In order to focus on the fall months, months were extracted from the index and .loc was used to select months greater than 8 and less than 12 in the O horizon. The index year was used from the fall seasons. Nan values were dropped again since we were dealing with original dataframe. By using the index year a groupby was used to find the mean values for each fall season throughout a decades worth of time. 

The streamwater data was conducted in a similar manner, but needed to use a .iloc to extract the fall months. 

```python
#Watershed 1 Soil Solution (O horizon)

#10 years time in the O horizon
startdate97 = pd.to_datetime('1997-9-01')
enddate07 = pd.to_datetime('2007-11-10') 

#Creating a new dataframe from the original
dfw1_soil_10 = dfw1_soil_og.copy()

#Trim OG df to Sept 1997 to Oct 2007
dfw1_soil_10 = dfw1_soil_10[startdate97:enddate07]

#Establishing fall months
dfw1_soil_10['MONTH'] = dfw1_soil_10.index.month
dfw1_soil_10fall = dfw1_soil_10.loc[(dfw1_soil_10['MONTH'] > 8) & (dfw1_soil_10['MONTH'] < 12)] 
dfw1_soil_10fall['YEAR'] = dfw1_soil_10fall.index.year
dfw1_soil_10fall = dfw1_soil_10fall.loc[(dfw1_soil_10fall['Hor'] == 'O')] 
dfw1_soil_10fall = dfw1_soil_10fall.dropna()

#Finding mean values for fall season
dfw1_soil_10fall_mean = dfw1_soil_10fall.groupby(['YEAR']).mean()
```

Finally, creating a plot based on the fall seasons for a decade's worth of time. Plotted is the Ca concentration and fluxes along silica's soil concentration vs the fall month years.

```python
def tenplot(df10, dfst):
    fig, (ax1) = plt.subplots(sharex = True)
    
    l1 = ax1.bar(df10.index, df10['Ca2_plus_'], color = 'c', label = 'Soil [Ca]')
    l2 = ax1.scatter(df10.index, df10['H4SiO4'], linestyle = '-', color = 'r', label = 'Soil [Si]')
    
    ax2 = ax1.twinx()
    l3 = ax2.plot(dfst['Ca_flux'],'g^-',label = 'Streamwater Ca')
    
    ax1.set_ylim(bottom = 0)
    ax1.set_ylim(top = 300)
    
    fig.autofmt_xdate()  
    
    ax1.set_ylabel('Elemental [] (ppb)', color='k')   
       
    ax1.set_xlabel('Time (Year)', color='k')
    
    ax2.set_ylim(bottom = 0, top = 2500)
    ax2.set_ylabel('Ca Flux (g/ha)', color='g')
    
    # Title the plot 
    ax1.set_title(' Annual [Ca] and [Si] Response Based on Fall Season from O Horizon')
    
    ax1.legend(loc = "center left", bbox_to_anchor=(-0.50, 0.05))
    ax2.legend(loc = "center left", bbox_to_anchor=(-0.50, 0.25))
    
    return df10, dfst

tenplot(dfw1_soil_10fall_mean, dfw1_stream_10fall_mean)
tenplot(dfw6_soil_10fall_mean, dfw6_stream_10fall_mean)
```

```python
# Plot O horizon soil Ca, pH, Time 

def CapH(df10):
    fig, ax1 = plt.subplots()
    line1 = ax1.plot(df10['Ca2_plus_'], color = 'r', marker = 'o', linestyle = '-', label = 'Ca Soil')    
        
    #Si line if wanted to see if pH plays a part..which it does not...
    # line3 = ax1.plot(dfmonthly_O10_2['H4SiO4'], color ='g', marker ='o', linestyle = '-', label = 'Si')
    
    ax1.set_ylabel('[Ca](ppb/season-year)', color='r')   
    
    for tl in ax1.get_yticklabels():
        tl.set_color('k')          
    
    ax1.set_ylim(bottom = 0)
    ax1.set_ylim(top = 140)
    
    fig.autofmt_xdate()                
    
    ax2 = ax1.twinx()
    line2 = ax2.plot(df10['pH'],'b^-',label = 'pH Soil')
    
    ax2.set_ylim(bottom = 3, top = 6)
    
    ax2.set_ylabel('pH (-log[H+])', color='b')    
    for t2 in ax2.get_yticklabels():
        t2.set_color('k')           
       
    ax1.set_title('Comparison of Soil [Ca] and pH ')
    
    lines = line1 + line2 
    labs = [l.get_label() for l in lines]
    ax1.legend(lines, labs, loc='best')
    
    plt.show()
    return df10

CapH(dfw1_soil_10fall_mean)
CapH(dfw6_soil_10fall_mean)
```
