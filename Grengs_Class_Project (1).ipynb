## 1. Introduction

Calcium (Ca) serves as a vital element as it is the most abundant mineral nutrient in Earth’s crust and plays an important role in chemical weathering especially within a forest (Likens et al., 1998; Hynicka et al., 2016). Often quantifying Ca is a challenge because of its mobility after being leached and liberated from soils, particularly in the organic (O) horizon (Federer et al., 1989; Bailey et al., 1996; Likens et al., 1998). As seen in the northeast, Ca depletion is often a consequence of acid rain and forest harvesting (Bailey et al., 1996; Blum et al., 2008; Green et al., 2013). The deposition of nitrogen and sulfur is also suggested to cause Ca leaching (Vadeboncoeur, 2010) along with other variables including low cation exchange capacities (CEC), mycorrhizal infections, tree frost injuries, and organic matter decomposition (Likens et al., 1998; Blum et al., 2008). Understanding how and when soil Ca concentrations are leached can help forest management practices make improvements for maintaining a healthy forest. Examples of these practices include adding fertilizers, liming, or adding organic matter (Sonon et al., 2017). 

In this study, two of Hubbard Brook Experimental Forest’s (HBEF) watersheds will be compared to observe when Ca values are at their highest within its watershed’s soil concentrations and streamwater fluxes. Hubbard Brook was an interesting site to select because in October of 1999 they had 45 tons of wollastonite (CaSiO3) added to one of its watersheds, specifically watershed 1 (W1) (Green et al., 2013; “Hubbard Brook Ecosystem Study”, n.d). Adding wollastonite to Hubbard Brook's W1 was in response to the acid deposition depleting the soil's Ca within the northeast region (Green et al., 2013). This analysis primarily focuses on when the soil's O horizon Ca concentrations peak in W1 and observe how adding wollastonite may affect the streamwater Ca flux.

The primary approach was first to establish the baseline Ca concentration and flux measurements to observe when Ca was at its highest without the effect of the wollastonite addition. It was assumed that baseline Ca concentrations and fluxes could be obtained from 2012 to 2017 by finding the averages for each month during this five-year span. This evaluation helped determine if there was a specific season in which Ca was highest within its O horizon, mineral (B) horizon, and streamwater chemistry. Based on when the O horizons Ca peak occurred, Ca was further focused on a decade timescale to show how the wollastonite addition may contribute to differences in streamwater, soil, and pH measurements. 

### **Specifically, the following questions will be addressed:**

* Within a five-year span, when will the soil's O horizon Ca concentration be highest seasonally at W1? Does the B horizon and streamwater follow suit? Will W6 show similarity with W1?


* If there is a seasonal pattern peak of the baseline Ca concentrations in the soil O horizon, how do Ca measurements change for a decade after the deposition of wollastonite in October 1999? Will the streamwater flux, soil O horizon silica concentration, and W6 follow a similar trend?


* Does Ca share a trend to any other potential variables like pH among the tested watersheds?




## 2. Site Description

The HBEF is located within the White Mountains National Forest in New Hampshire at 43.95°N and 71.78°W (Remiszewski et al., 2016). This site was established in 1955 by the United States Forest Service and serves as an important location to conduct research for long-term projects specializing in water, plants, air, animals, and soils. The overall size of Hubbard Brook is approximately 3,200 hectares. Hubbard Brook is part of the Merrimack River watershed where it serves as a tributary to the Pemigewasset River (Campbell et al., 2020). It also has a humid continental climate where the mean annual temperature is 5.5 °C and mean annual precipitation is approximately 1400 mm (Holmes and Likens, 1998; Remiszewski et al., 2016; Campbell et al., 2020; “Hubbard Brook Ecosystem Study”, n.d.). Primarily, Hubbard Brook's soil are spodosols which is a product from the glacial till present there (Cho et al., 2010). Hubbard Brook's elevational differences play a role in the soil's thickness. At higher elevations, soil's depths are shallow and associated with the bedrock outcrops; whereas, lower elevations have more glacier till and are deeper in depth (Cho et al., 2010). Hubbard Brook's primary bedrock formations are dominated by the Rangeley Formation, Kinsman Granodiorite, Littleton Formation, and Perry Mountain Formation. These rock formations are known to be composed of schist, Concord granites, pegmatites, tonalite, diabase, and quartzite (Lyons et al., 1997). 

As mentioned previously, I focused on Hubbard Brook’s W1 and W6 to investigate and compare when Ca concentrations differ among each watershed’s soil and streamwater Ca fluxes (Figure 1). Watershed 1 was selected because it received a wollastonite treatment in October of 1999 and W6 was selected to serve as a reference. Although these watersheds are located within the same area, they share some similarities and differences between area, slope, elevation, treatment, soil type(s), and vegetation (Table 1).

### 2.1 Datasets Utilized

Two datasets were used to observe when W1 and W6 Ca measurements were at their peak. As seen in Table 2, the datasets included each watershed’s measurements for streamwater chemistry fluxes and soil solution chemistry; along with its respected dates, data frequency, and creator. The data was imported based on the recommendations from the creators for a smooth transition into python. The imported data was modified by creating functions for both the soil solutions and streamwater chemistry datafiles. This suggestion and function helped streamline each watershed's data quicker and more efficiently. 

For the soil solutions data, a dictionary was created to store simple soil horizons (simple_hori). Sub-horizons Oa, Oe, and Oae were grouped into a simple O horizon group. Along with, sub-horizons Bh, Bhs, Bhs1, Bhs3, Bs, and Bs2 were grouped into a simple B horizon group. Doing this type of grouping was also done in past studies. It was also assumed that by grouping the sub-horizons together that this would fill in any missing values since they all shared common sub-horizon values. The soil solutions data was filtered to bring in the following columns: watershed, horizon, Ca concentration, H4SiO4 (Si) concentration, date, and pH. The index was set to the date. 

The streamwater chemistry data presented a different challenge since the exact dates was not recorded like the soil solutions data. When importing began, the index was set to year and was later re-visited as discussed later. The streamwater data was filtered to bring in the following columns: month, flow, Ca flux, SiO2 flux, year datetime, pH, and year-month.

Both datasets NA values were set to equal -999.99, -888.88, -777.77, 'NaN', 'Ice', or 'Eqp'. 


```python
#To import data, using what HB creators recommended for smooth transition
filenameW1_soil  ="https://pasta.lternet.edu/package/data/eml/knb-lter-hbr/138/9/e531788965a5622a4ca8e664a9d06406".strip() 
filenameW1_soil  = filenameW1_soil.replace("https://","http://")
filenameW6_soil  ="https://pasta.lternet.edu/package/data/eml/knb-lter-hbr/62/16/4fa59538abe2e6ff15a35ea2de913546".strip() 
filenameW6_soil  = filenameW6_soil.replace("https://","http://")
filenameW1_stream  ="https://pasta.lternet.edu/package/data/eml/knb-lter-hbr/3/17/520d38828fe2356314e51008a5059dd4".strip() 
filenameW1_stream  = filenameW1_stream.replace("https://","http://")
filenameW6_stream  ="https://pasta.lternet.edu/package/data/eml/knb-lter-hbr/8/17/3312389e77cc5fd06bc8a7c9019de0ed".strip() 
filenameW6_stream  = filenameW6_stream.replace("https://","http://")

#Created a dictoriary before importing soils data
simple_hori = {'Oa' : 'O', 'Oe' : 'O', 'Oae' : 'O', 'Bh' : 'B', 'Bhs' : 'B', 'Bhs1' : 'B', 'Bhs3' : 'B', 'Bs' : 'B', 'Bs2' : 'B'}

#Read in soil chemistry datafile
def read_in_soils(file):
    df = pd.read_csv(file, skiprows=1, sep=",", names=["date","WS","Elev","Hor","pH","ANC","H4SiO4","Na_plus_",     
             "K_plus_","Ca2_plus_","CaFLAG", "Mg2_plus_","Alt","Alm", "Alo","NH4_plus_", "F_hyphen_",     
             "Cl_hyphen_", "SO4=","NO3_hyphen_","DIC", "DOC","TN"],
             parse_dates=['date'], na_values = [-999.99,-888.88, -777.77, 'NaN', 'Ice', 'Eqp']) 
    df=df.assign(date_datetime=pd.to_datetime(df.date,errors='coerce'))  
    df.WS=df.WS.astype('category')  
    df.Elev=df.Elev.astype('category')  
    df.Hor=df.Hor.astype('category') 
    df.pH=pd.to_numeric(df.pH,errors='coerce') 
    df.ANC=pd.to_numeric(df.ANC,errors='coerce') 
    df.H4SiO4=pd.to_numeric(df.H4SiO4,errors='coerce') 
    df.Na_plus_=pd.to_numeric(df.Na_plus_,errors='coerce') 
    df.K_plus_=pd.to_numeric(df.K_plus_,errors='coerce') 
    df.Ca2_plus_=pd.to_numeric(df.Ca2_plus_,errors='coerce')  
    df.CaFLAG=df.CaFLAG.astype('category') 
    df.Mg2_plus_=pd.to_numeric(df.Mg2_plus_,errors='coerce') 
    df.Alt=pd.to_numeric(df.Alt,errors='coerce') 
    df.Alm=pd.to_numeric(df.Alm,errors='coerce') 
    df.Alo=pd.to_numeric(df.Alo,errors='coerce') 
    df.NH4_plus_=pd.to_numeric(df.NH4_plus_,errors='coerce') 
    df.F_hyphen_=pd.to_numeric(df.F_hyphen_,errors='coerce') 
    df.Cl_hyphen_=pd.to_numeric(df.Cl_hyphen_,errors='coerce') 
    # df.SO4=pd.to_numeric(df.SO4=,errors='coerce') 
    df.NO3_hyphen_=pd.to_numeric(df.NO3_hyphen_,errors='coerce') 
    df.DIC=pd.to_numeric(df.DIC,errors='coerce') 
    df.DOC=pd.to_numeric(df.DOC,errors='coerce') 
    df.TN=pd.to_numeric(df.TN,errors='coerce') 
    df = df.set_index(['date'])
    df = df[['WS', 'Hor','Ca2_plus_', 'H4SiO4', 'date_datetime', 'pH']]
    # df = df[['WS', 'Hor','Ca2_plus_', 'H4SiO4', 'date_datetime', 'pH', 'Mg2_plus_', 'K_plus_', 'Na_plus_',
    #          'Alo', 'NH4_plus_']]
    df = df.replace({"Hor" : simple_hori})
    
    return df

#Read in streamwater chemistry datafile
def read_in_streamwater(file):
    df =pd.read_csv(file ,skiprows=1 ,sep=",",quotechar='"', names=["Year", "Month","Year_Month", "flow_mm", "Ca_flux","Mg_flux",     
            "K_flux", "Na_flux", "Al_Ferron_flux", "TMAl_flux", "OMAl_flux","Al_ICP_flux","NH4_flux","SO4_flux",     
            "NO3_flux","Cl_flux","PO4_flux","DOC_flux", "TDN_flux","DON_flux", "DIC_flux","SiO2_flux","Mn_flux",     
            "Fe_flux", "F_flux","H_flux", "pH_volwt", "SpecCond_volwt","ANC_volwt"],
            parse_dates=['Year'], na_values = [-999.99,-888.88, -777.77, 'NaN', 'Ice', 'Eqp']) 
    df=df.assign(Year_datetime=pd.to_datetime(df.Year,errors='coerce'))  
    df.Month=df.Month.astype('category')  
    df.Year_Month=df.Year_Month.astype('category') 
    df.flow_mm=pd.to_numeric(df.flow_mm,errors='coerce') 
    df.Ca_flux=pd.to_numeric(df.Ca_flux,errors='coerce') 
    df.Mg_flux=pd.to_numeric(df.Mg_flux,errors='coerce') 
    df.K_flux=pd.to_numeric(df.K_flux,errors='coerce') 
    df.Na_flux=pd.to_numeric(df.Na_flux,errors='coerce') 
    df.Al_Ferron_flux=pd.to_numeric(df.Al_Ferron_flux,errors='coerce',downcast='integer') 
    df.TMAl_flux=pd.to_numeric(df.TMAl_flux,errors='coerce') 
    df.OMAl_flux=pd.to_numeric(df.OMAl_flux,errors='coerce') 
    df.Al_ICP_flux=pd.to_numeric(df.Al_ICP_flux,errors='coerce') 
    df.NH4_flux=pd.to_numeric(df.NH4_flux,errors='coerce') 
    df.SO4_flux=pd.to_numeric(df.SO4_flux,errors='coerce') 
    df.NO3_flux=pd.to_numeric(df.NO3_flux,errors='coerce') 
    df.Cl_flux=pd.to_numeric(df.Cl_flux,errors='coerce') 
    df.PO4_flux=pd.to_numeric(df.PO4_flux,errors='coerce') 
    df.DOC_flux=pd.to_numeric(df.DOC_flux,errors='coerce') 
    df.TDN_flux=pd.to_numeric(df.TDN_flux,errors='coerce') 
    df.DON_flux=pd.to_numeric(df.DON_flux,errors='coerce') 
    df.DIC_flux=pd.to_numeric(df.DIC_flux,errors='coerce') 
    df.SiO2_flux=pd.to_numeric(df.SiO2_flux,errors='coerce') 
    df.Mn_flux=pd.to_numeric(df.Mn_flux,errors='coerce') 
    df.Fe_flux=pd.to_numeric(df.Fe_flux,errors='coerce') 
    df.F_flux=pd.to_numeric(df.F_flux,errors='coerce') 
    df.H_flux=pd.to_numeric(df.H_flux,errors='coerce') 
    df.pH_volwt=pd.to_numeric(df.pH_volwt,errors='coerce') 
    df.SpecCond_volwt=pd.to_numeric(df.SpecCond_volwt,errors='coerce') 
    df.ANC_volwt=pd.to_numeric(df.ANC_volwt,errors='coerce') 
    df = df.set_index(['Year'])
    df = df[['Month', 'flow_mm', 'Ca_flux', 'Year_datetime', 'pH_volwt', 'Year_Month', 'SiO2_flux']]

    return df
```

```python
#File reading for each watershed's soil and streamwater chemistry
dfW1_soil = read_in_soils(filenameW1_soil)
dfW6_soil = read_in_soils(filenameW6_soil)
dfW1_stream = read_in_streamwater(filenameW1_stream)
dfW6_stream = read_in_streamwater(filenameW6_stream)
```

```python
#Created a copy of the original (og) dataset
dfw1_soil_og = dfW1_soil.copy()
dfw6_soil_og = dfW6_soil.copy()
dfw1_streams_og = dfW1_stream.copy()
dfw6_streams_og = dfW6_stream.copy()
```

### 3.2 Finding Calcium Baseline Seasonality

The soil solutions datasets were further filtered on by selecting a five-year span that wasn't affected by CaSiO3 addition. Selecting and trimming the dates from December 2012 to December 2017 helped establish baseline Ca and Si soil concentrations for comparison between W1 and W6. The NA values were dropped since the sub-horizons were grouped into larger, simpler horizons as previously mentioned, and was assumed that this would lessen variability among the data. To make calculations easier, each watershed's year, month, and day was extracted from the index.

New dataframes were created to get the soil O and B horizons separated for timeseries plotting. From those dataframes, the mean values were calculated from each month within the five-year span. Originally, the  soil O and B horizon dataframes were going to be merged; however, problems in python caused the O horizon to average values. Thus, it was decided to keep each watershed's soil horizons in separate dataframes for this part of the analysis.

```python
#Selecting a five year span to establish a baseline Ca concentration
startdate5_soil = pd.to_datetime('2012-12-14')
enddate5_soil = pd.to_datetime('2017-12-17') 
```

```python
#Triming the dataframes to data that occur between the desired baseline start date and end date
dfW1_soil = dfW1_soil[startdate5_soil:enddate5_soil]
dfW6_soil = dfW6_soil[startdate5_soil:enddate5_soil]

dfW1_soil = dfW1_soil.dropna()
dfW6_soil = dfW6_soil.dropna()

dfW1_soil['YEAR'] = dfW1_soil.index.year
dfW1_soil['DAY'] = dfW1_soil.index.dayofyear
dfW1_soil['MONTH'] = dfW1_soil.index.month

dfW6_soil['YEAR'] = dfW6_soil.index.year
dfW6_soil['DAY'] = dfW6_soil.index.dayofyear
dfW6_soil['MONTH'] = dfW6_soil.index.month

# Separting O and B horizons
dfW1_soil_O = dfW1_soil.loc[(dfW1_soil['Hor'] == 'O')]
dfW6_soil_O = dfW6_soil.loc[(dfW6_soil['Hor'] == 'O')]
dfW1_soil_B = dfW1_soil.loc[(dfW1_soil['Hor'] == 'B')]
dfW6_soil_B = dfW6_soil.loc[(dfW6_soil['Hor'] == 'B')]

#Merge them causes the O horizon to average (or do something funky) right away
# dfW1_soil = dfW1_soil_O.merge(dfW1_soil_B, how = 'inner', left_index = True, right_index = True) 
# dfW6_soil = dfW6_soil_O.merge(dfW6_soil_B, how = 'inner', left_index = True, right_index = True)

#Finding mean values for horizons
dfW1_soilO_mean = dfW1_soil_O.groupby(['MONTH']).mean()
dfW6_soilO_mean = dfW6_soil_O.groupby(['MONTH']).mean()
dfW1_soilB_mean = dfW1_soil_B.groupby(['MONTH']).mean()
dfW6_soilB_mean = dfW6_soil_B.groupby(['MONTH']).mean()
```

A similar approach was taken with the streamwater chemistry data by trimming the years from Decemeber 2012 to 2017. A month column was already established within the datafile, hence mean values were easily computed for each month within the five year span. 

```python
#Streamwater
dfW1_stream['YEAR'] = dfW1_stream.index.year
dfW6_stream['YEAR'] = dfW6_stream.index.year

#Trim years
start_yr_stm = pd.to_datetime('2012')
end_yr_stm = pd.to_datetime('2017')
dfW1_stream = dfW1_stream[start_yr_stm:end_yr_stm]
dfW6_stream = dfW6_stream[start_yr_stm:end_yr_stm]

#Finding mean values for watersheds
dfW1_stream_mean = dfW1_stream.groupby(['Month']).mean()
dfW6_stream_mean = dfW6_stream.groupby(['Month']).mean()
```

From the revised datasets, a function was created to display each watershed's baseline Ca concentrations (Figure 2). As shown in Figure 2, the first plot is showing W1 and then W6. The blue line symbolizes the O horizon, red line is B horizon, and green is the streamwater. Note that the soil's data is units of parts per billion (ppb) and streamwater fluxes are in grams per hectare (g/ha). Months are labelled numercially along the x-axis. Please note that for the remaining plots that W1 will display first and then W6. Plots will be discussed in results section.

```python
def timeseriesplot(meanO, meanB, meanS):
    fig, (ax1) = plt.subplots(sharex = True)
    
    l1 = ax1.plot((meanO['Ca2_plus_']), color = 'mediumseagreen', linestyle = '-', label = 'O Horizon')
    l2 = ax1.plot((meanB['Ca2_plus_']), color = 'deeppink', linestyle = '-', label = 'B Horizon')
    
    ax2 = ax1.twinx()
    l5 = ax2.plot(meanS['Ca_flux'], color = 'navy', marker = '1',label = 'Streamwater')
    
    ax2.set_ylim(bottom = 0, top = 3000)
    ax2.set_ylabel('Streamwater Ca Flux (g/ha)', color='navy')
    
    fig.autofmt_xdate() 
    
    ax1.set_xlabel('Time (Month #)', color='k')
    ax1.set_ylabel('Soil Horizon [Ca] Monthly Mean (ppb)', color='k')   

    ax1.set_ylim(bottom = 0)
    ax1.set_ylim(top = 120)

    ax1.set_title("Baseline Ca Levels During 2012-2017", fontweight="bold")
    
    lines = l1 + l2 + l5 
    labs = [l.get_label() for l in lines]
    ax1.legend(lines, labs, loc=(1.1, 1.0))
    
    return meanO, meanB, meanS

timeseriesplot(dfW1_soilO_mean, dfW1_soilB_mean, dfW1_stream_mean)
timeseriesplot(dfW6_soilO_mean, dfW6_soilB_mean, dfW6_stream_mean)
```

### 3.3 Soil and Streamwater Responses Pre- and Post Wollastonite Addition

In the next section, a decade's worth of time from September 1997 to November 2007 was used by specifically focussing on the fall months of September, October, and November. The fall season was selected since  W1's baseline soil O horizon peaked in Ca concentrations during these months and was assumed less variability would occur if one was only to focus on a single month. By increasing time and placing a focus on this season, observations were made to demonstrate how the wollastonite addition affected Ca and Si measurements before and after the treatment. It should be noted that the wollastonite addition is referred to as the Ca addition within the plots. 

To do this, a new dataframe was created from the soil's original dataframe and had it trimmed from dates ranging from  September 1997 to November 2007. To focus on the fall months, months were extracted from the index and .loc was used for selecting months greater than 8 and less than 12 in the soil O horizon. The index year was used from the fall seasons. NA values were dropped again since data was wrangled by the simple horizon groupings that were used earlier with the original dataframe. By using the index year a groupby was used to find the mean values for each fall season throughout a decade worth of time. 

The streamwater data was conducted in a similar manner but needed to use a .iloc to extract the fall months.  

```python
#Watershed 1 Soil Solution (O horizon)

#10 years time in the O horizon
startdate97 = pd.to_datetime('1997-9-01')
enddate07 = pd.to_datetime('2007-11-10') 

#Creating a new dataframe from the original
dfw1_soil_10 = dfw1_soil_og.copy()

#Trim OG df to Sept 1997 to Oct 2007
dfw1_soil_10 = dfw1_soil_10[startdate97:enddate07]

#Establishing fall months
dfw1_soil_10['MONTH'] = dfw1_soil_10.index.month
dfw1_soil_10fall = dfw1_soil_10.loc[(dfw1_soil_10['MONTH'] > 8) & (dfw1_soil_10['MONTH'] < 12)] 
dfw1_soil_10fall['YEAR'] = dfw1_soil_10fall.index.year
dfw1_soil_10fall = dfw1_soil_10fall.loc[(dfw1_soil_10fall['Hor'] == 'O')] 
dfw1_soil_10fall = dfw1_soil_10fall.dropna()

#Finding mean values for fall season
dfw1_soil_10fall_mean = dfw1_soil_10fall.groupby(['YEAR']).mean()
```

Finally, W1 and W6 figures were created based on the fall seasons for a decade's worth of time by using a function for plotting. Plotted is the Ca concentrations and fluxes along with Si's concentration and fluxes vs the fall season-years (Figure 3). 

```python
def tenplot(df10, dfst):
    fig, (ax1) = plt.subplots(sharex = True)
    
    l1 = ax1.bar(df10.index, df10['Ca2_plus_'], color = 'mediumseagreen', label = 'Soil [Ca]')
    l2 = ax1.scatter(df10.index, df10['H4SiO4'], linestyle = '-', color = 'coral', label = 'Soil [Si]')
    
    ax2 = ax1.twinx()
    l3 = ax2.plot(dfst['Ca_flux'],color = 'navy', marker = '1',label = 'Streamwater Ca')
    l4 = ax2.plot(dfst['SiO2_flux'],color = 'mediumblue', marker = '1', label = 'Streamwater Si')
    
    ax1.set_ylim(bottom = 0)
    ax1.set_ylim(top = 300)
    
    fig.autofmt_xdate()  
    
    ax1.set_ylabel('Soil Elemental [] (ppb)', color='k')   
       
    ax1.set_xlabel('Time (Fall Season-Year)', color='k')
    
    ax2.set_ylim(bottom = 0, top = 10000)
    ax2.set_ylabel('Streamwater Elemental Flux (g/ha)', color='midnightblue')
    
    plt.axvline(x=  1999)
    ax1.text(1999, 200, "Ca Addition", ha='center', fontweight="bold")
    
    # Title the plot 
    ax1.set_title(' Annual [Ca] and [Si] Response Based on Fall Season from O Horizon', fontweight="bold")
    
    ax1.legend(loc = "center left", bbox_to_anchor=(-0.50, 0.05))
    ax2.legend(loc = "center left", bbox_to_anchor=(-0.50, 0.25))
    
    plt.show
    
    return df10, dfst

tenplot(dfw1_soil_10fall_mean, dfw1_stream_10fall_mean)
tenplot(dfw6_soil_10fall_mean, dfw6_stream_10fall_mean)
```

### 3.4 Timeseries of Soil and Streamwater Calcium vs pH

Further timeseries analysis was conducted to observe any similar trends between Ca and pH among the test watersheds. A plotting function was created to compare the soil O horizon's pH levels and Ca concentrations (Figure 4), and the streamwater's Ca flux with pH (Figure 5). These plots were also based on the same decade timescale as previously used. For each cycle of plots, W1 was calcualted first and then W6.

```python
# Plot O horizon soil Ca, pH, Time 

def CapH(df10):
    fig, ax1 = plt.subplots()
    line1 = ax1.plot(df10['Ca2_plus_'], color = 'forestgreen', marker = '2', linestyle = '-', label = 'Soil [Ca]')    
        
    #Si line if wanted to see if pH plays a part..which it does not...
    # line3 = ax1.plot(dfmonthly_O10_2['H4SiO4'], color ='g', marker ='o', linestyle = '-', label = 'Si')
    
    ax1.set_ylabel('[Ca](ppb/season-year)', color='forestgreen')   
    
    for tl in ax1.get_yticklabels():
        tl.set_color('k')          
    
    ax1.set_ylim(bottom = 0)
    ax1.set_ylim(top = 150)
    
    fig.autofmt_xdate()                
    
    ax2 = ax1.twinx()
    line2 = ax2.plot(df10['pH'],color = 'darkred', marker = '1',label = 'pH Soil')
    
    ax2.set_ylim(bottom = 3, top = 6)
    
    ax2.set_ylabel('pH (-log[H+])', color='darkred')    
    for t2 in ax2.get_yticklabels():
        t2.set_color('k') 
        
    plt.axvline(x=  1999)
    ax1.text(1999, 120, "Ca Addition", ha='center', fontweight="bold")
    ax1.set_xlabel('Time (Fall Season-Year)', color='k')
       
    ax1.set_title('Comparison of Soil O Horizon [Ca] and pH', fontweight="bold")
    
    lines = line1 + line2 
    labs = [l.get_label() for l in lines]
    ax1.legend(lines, labs, loc='best')
    
    plt.show()
    return df10

CapH(dfw1_soil_10fall_mean)
CapH(dfw6_soil_10fall_mean)
```

## 5. Discussion and Conclusions 


### 5.1 Potential Causes of Escalated Baseline Ca Concentrations
The analyses demonstrated how the wollastonite addition has continued to affect the soil O horizon Ca concentrations and the seasonal cycling of Ca in W1 in comparison to W6. In the soil O horizon, Ca was most abundant in the fall (Figures 1 and 6). This trend may suggest when leaves are falling off trees and thus having leaf litter decomposing into the soil’s organic layer to elevate Ca (Dijkstra and Smits, 2002). The different tree species surrounding each watershed’s lysimeter sampling location could also affect the soil's Ca concentrations by their foliage and/or root system too. In Dijkstra and Smits (2002) budget analysis, sugar maples were shown to have higher Ca concentrations because of their deep root network in comparison to hemlocks displaying the opposite effect. However, greenhouse experiments have shown consideration needs to include root traits like age, water influx, and fungi associations (Dauer et al., 2007). Needless to say, a stronger case could be developed if specific tree species were considered and partitioned. 

The soil B horizon and streamwater fluxes displayed similar trends between W1 and W6. In both W1 and W6, the Ca concentration in the soil B horizon was pretty consistent and can provide further evidence on Ca’s retention in the soil O horizon (Figures 1 and 6). The behavior suggests that Ca concentrations have decreased with increasing soil depth (Cho et al., 2010) for increased retention in the organic layer. The streamwater Ca flux also demonstrated a strong seasonality but in the spring (Figures 1 and 6). This trend was attributed to a larger flux due to the vernal window and appeared that the streamwater Ca flux does not seem to be influenced by the wollastonite addition. Precipitation probably played a larger role in the streamwater's flux measurements and further analysis of this speculation would need to be done for confirmation. 

### 5.2 Cation Exchange Playing A Role
During the fall, the baseline Ca concentrations peaked in W1 which then allowed further analyses on how Ca responded pre-and post-treatment. As presented in Figure 7, Ca substantially increased and peaked in 2001; whereas, Si increased the following year in 2000. Silica concentrations likely increased quicker because it was more readily dissolved or uptake in vegetation (Cho et al., 2012). The increase in Si concentrations did not seem to affect the Si flux measurements in streamwater. The two-year lag in peak Ca concentrations may be attributed to the soil’s cation exchange capacity or vegetation’s initial uptake of Ca (Green et al., 2013). It is reasonable that the CEC increased along with pH increasing (Figure 7). Cation exchange capacity occurs when soils hold a more negative charge for base cations such as Ca, magnesium, and potassium to bond with (Ma and Eggleton, 1999). Clayey soils have a more negative charge, like kaolinites, have been shown to increase in cation exchange capacity and directly correlate with pH (Ma and Eggleton, 1999). Kaolinite was saturated within this timeframe (Cho et al., 2012) and further supports the idea of increasing CEC in HBEF soils along with its increase of pH. 

### 5.3 Improvements
To further speculate on the  soil O horizon Ca concentration, more improvements could be implemented in the field and through statistical testing. First, more lysimeter sampling sites could be installed in W6. Currently, there was only 3 sites in comparison to W1’s 13. By increasing the number of sampling sites, it could provide a better representation of W6 as a control watershed. Likewise, conducting a statistical test, such as an ANOVA, would strengthen the argument made for treads seen in Ca concentrations and pH by comparing their variances to one another.

In this study, more variables could have been beneficial to include. Elevation could have played a larger role since it has been mentioned within a previous literature that lower elevations tended to have more glacial till in comparison to higher elevations having shallow amounts (Cho et al., 2012). However, it was suggested in much of the literature that Ca was largest retained within the soil’s O horizon and increased CEC across all elevations (Cho et al., 2010; Cho et al., 2012). It would have also been beneficial to obtain the soil depths at each sampling location to help provide a better idea of how each of the watershed’s Ca cycles further differed. As suggested earlier in Dijkstra and Smits (2002) and Dauer et al. (2007), Ca seems to be partitioned in different tree species and is likely to affect their Ca cycles. 

Quantitative calculations of variables computed to strengthen observed trends. Calculated variables like base saturation and CEC could have been made for further comparisons between the watersheds along wtih other sites in the United States. It would have also been interesting to see how these variables could have changed throughout time. Along with those calculations, the mineral saturation of wollastonite could have been calculated by using software programs like Geochemist Workbench or Visual MINTEQ and imported into python plot changes. By calculating wollastonite's saturation, one could observe how long Ca was retained in the soil and interpolate to see how long it last before it completely weathers out.

### 5.4 Conclusion
In closing, a few speculations were made. In W1, soil O horizon Ca concentrations peaked in the fall, whereas W6 did not show a similar trend. This behavior was likely demonstrating how wollastonite was continuing to play a role in the soil after 13 to 18 years. In both watersheds, soil B horizon did not show much variability; however, the streamwater fluxes had peak Ca flux in spring due to snowmelt. On a decade timescale, it was interesting to note that Ca concentrations did not initially peak after the wollastonite application showing that it may be immobilized at first before being fully retained within the soil. The Ca concentrations and pH peaked in the fall of 2001. The Ca concentrations and pH was strong enough to speculate how the CEC was potentially increasing with increasing Ca concentrations and pH in W1. Watershed 6 also showed a similar trend but not as prominent. Overall, this analysis showed how adding wollastonite continues to affect soil concentrations and pH levels to lower the acidity levels for increased tree growth and lower deficiencies. Although this study only focused on Hubbard Brook but could be further improved on and then compared to other northeastern forest sites.
